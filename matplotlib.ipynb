```python
import numpy as np
import matplotlib.pyplot as plt

x = np.array(["S1","S2","S3","S4","S5","S6","S7"])
y = np.array([75, 80, 83, 97, 90, 93, 89])

plt.title("Subject marks")
plt.xlabel("subject name")
plt.ylabel("marks")

plt.plot(x, y)

plt.grid()

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array([30,20,10,25,15,4,1])
y = ["Aerobics","Soccer","Swimming","Walking","Jogging","Yoga","Bicycling"]
plt.pie(x,labels=y)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array([30,20,21,25,15,42,10])
y = ["Aerobics","Soccer","Swimming","Walking","Jogging","Yoga","Bicycling"]
z = [0.4,0,0,0,0,0,0.4]
c = ["MediumSeaGreen","AliceBlue","#F0FFFF","Cornsilk","FloralWhite","#FFFFF0","IndianRed"]
plt.pie(x,labels=y,explode=z,shadow=True,startangle=90,colors=c)
plt.legend(title="Hobbies")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

#plot 1:
x = np.array(["S1","S2","S3","S4"])
y = np.array([84,87,85,90])

plt.subplot(1, 2, 1)
plt.plot(x,y)
plt.title("Sem1")

#plot 2:
x = np.array(["S1","S2","S3","S4"])
y = np.array([93,90,89,95])

plt.subplot(1, 2, 2)
plt.plot(x,y)
plt.title("sem2")

plt.show()
```
